```python
def categorize_experience(experience):
    if experience < 5:
        return "less than 5 years"
    elif experience < 10:
        return "5-10 years"
    elif experience < 15:
        return "10-15 years"
    elif experience < 20:
        return "15-20 years"
    elif experience < 25:
        return "20-25 years"
    else:
        return "25+ years"
```

```python
df['experience_category'] = df['experience_coding'].apply(categorize_experience)
```

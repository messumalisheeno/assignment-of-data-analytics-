1. List all cities where the area is greater than 5000, along with their population.

```python
q = '''select c.name city,f.area area,c.population population  from facts f
    inner join cities c
    on 
    f.id = c.facts_id
    where
    f.area > 5000
'''
sql_query(q)
```

2-List the names of capital cities where the death rate is greater than 2% and the
birth rate is less than 20%

```python
q ='''select c.name capital_city,f.death_rate death_rate,f.birth_rate birth_rate from facts f
   inner join cities c
   on
   f.id = c.facts_id
   where c.capital = 1
   and f.death_rate > 2
   and f.birth_rate < 20
'''
sql_query(q)
```

3. List the cities with a population growth rate between 1% and 5%, sorted in desc
by population growth.

```python
q ='''select c.name city,f.population_growth growth_rate from facts f
   inner join cities c
   on
   f.id = c.facts_id
   where f.population_growth between 1 and 5
   order by population_growth desc
'''
sql_query(q)
```

4. Find countries that start with &quot;A&quot; and do not have any cities listed in the cities
table. The query should return the country name and city name. Using left join

```python
q = '''select f.name country,c.name city from facts f
    left join cities c
    on
    f.id = c.facts_id
    where 
    f.name like 'A%' and c.name is null
'''
sql_query(q)
```

5. List all capital cities and their corresponding death rate using a LEFT JOIN.

```python
q = '''select c.name capital_city,f.death_rate death_rate from facts f 
    left join cities c
    on
    f.id = c.facts_id
    where c.capital = 1
'''
sql_query(q)
```

5. Are there more male or female customers, Show in Pie Chart?

```python
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6, 6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90)
plt.title("Distribution of Male and Female Customers")
plt.ylabel('') 
plt.show()
```

6. Which category is least popular among customers?

```python
category_counts = df['category'].value_counts()
least_popular_category = category_counts.idxmin()
least_popular_count = category_counts.min()

print(f"The least popular category is '{least_popular_category}' with {least_popular_count} customers.")
```
